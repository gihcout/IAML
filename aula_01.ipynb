Machine learning - Regressão Linear

```python
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split

# carrega a base de dados
X, y = load_diabetes(return_X_y=True)
X = X[:, [2]]  # Use only one feature
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=20, shuffle=False)
```

```python
from sklearn.linear_model import LinearRegression

# regressão linear é realizada apenas nessa linha
regressor = LinearRegression().fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

y_pred = regressor.predict(X_test)

print(f"Mean squared error: {mean_squared_error(y_test, y_pred):.2f}")
print(f"Coefficient of determination: {r2_score(y_test, y_pred):.2f}")
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(ncols=2, figsize=(10, 5), sharex=True, sharey=True)

ax[0].scatter(X_train, y_train, label="Train data points")
ax[0].plot(
    X_train,
    regressor.predict(X_train),
    linewidth=3,
    color="tab:orange",
    label="Model predictions",
)
ax[0].set(xlabel="Feature", ylabel="Target", title="Train set")
ax[0].legend()

ax[1].scatter(X_test, y_test, label="Test data points")
ax[1].plot(X_test, y_pred, linewidth=3, color="tab:orange", label="Model predictions")
ax[1].set(xlabel="Feature", ylabel="Target", title="Test set")
ax[1].legend()

fig.suptitle("Linear Regression")

plt.show()
```

Case 02 - Dinheiro traz felicidade?
https://www.perplexity.ai/page/dinheiro-e-felicidade-uma-anal-LAGHzLkvTD2RRqpXkxytlA

```python
from sklearn.model_selection import train_test_split
import numpy as np

#criando os dados
x = np.array([5000,  20000, 52500, 137500, 350000, 500000]).reshape(-1, 1) 
y = np.array([4.0, 5.2, 6.8, 7.5, 8.1, 8.6])

# Criação e treinamento do modelo de regressão linear
modelo = LinearRegression()
modelo.fit(x, y)
```

```python
# Previsões do modelo
felicidade_pred = modelo.predict(x)
```

```python
# Cálculo do R²
r2 = r2_score(y, felicidade_pred)
```

```python
# Configurações do gráfico
plt.figure(figsize=(12, 7))
plt.scatter(x/1000, y, color='darkblue', s=100, label='Dados Observados')
plt.plot(x/1000, felicidade_pred, color='red', linewidth=2,
        label=f'Regressão Linear\n(y = {modelo.coef_[0]:.5f}x + {modelo.intercept_:.2f})')

plt.title('Relação entre Renda e Felicidade', fontsize=14, fontweight='bold')
plt.xlabel('Renda Anual (milhares de USD)', fontsize=12)
plt.ylabel('Índice de Felicidade (0-10)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.text(400, 4.5, f'R² = {r2:.3f}', fontsize=12, bbox=dict(facecolor='white', alpha=0.9))

# Ajustes de escala para melhor visualização
plt.xlim(0, 550)
plt.ylim(3, 10)

plt.show()
```
